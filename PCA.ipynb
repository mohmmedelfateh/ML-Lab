<span style="font-family: Comic Sans MS;font-size: 3em;color: #159895; background: inherit;">
    Covariance Matrix
    </span>

<span style="font-family: Comic Sans MS;font-size: 1.5em;color: #1A5F7A; background: inherit;">
Well need to first understand what is variance and covariance.
<ul>
  <li>Variance: measures how much your data is spread out. </li>
  <li>Covariance : measure that describes the relationship between variables(correlation).</li>
</ul>
</span>

<span style="font-family: Comic Sans MS;font-size: 1.5em;color: #1A5F7A; background: inherit;">
Covariance Matrix:
    <br /><br />
    It is a square matrix that shows the variance of variables and the covariance between a pair of variables in a dataset.
    <br /><br />
     X, Y, and Z with their values as shown below then we would first calculate the variance of X, Y, and Z which is 80.3, 33.037, and 142.5 respectively.
</span>

<span style="font-family: Comic Sans MS;font-size: 1.5em;color: #1A5F7A; background: inherit;">
cov(X,Y) is -13.865  This is a negative number which means as X increases, Y decreases, and vice-versa.
        <br /><br />
cov(X,Z) is 14.25  This is a positive number which means as X increases Z increases, and vice-versa.
        <br /><br />
cov(Y,Z) is -39.525  This is a negative number which means as Y increases, Z decreases and vice-versa.
</span>

 <span style="font-family: Comic Sans MS;font-size: 1.5em;color: #1A5F7A; background: inherit;">
How Did It Work? Scan That

 <span style="font-family: Comic Sans MS;font-size: 1.5em;color: #1A5F7A; background: inherit;">
To compute the % variance preserved by each component, we divide the Eigenvalue of each component by the sum of all Eigen Values.
</span>

<span style="font-family: Comic Sans MS;font-size: 3em;color: #159895; background: inherit;">
    Feature Vector or Matrix
    </span>

 <span style="font-family: Comic Sans MS;font-size: 1.5em;color: #1A5F7A; background: inherit;">
A feature vector is a matrix of all the Eigenvectors which we have decided to keep column-wise.
If we choose to keep m Eigenvectors out of n, the final data will have m dimensions.</span>

<span style="font-family: Comic Sans MS;font-size: 3em;color: #159895; background: inherit;">
Align/ Recast data along Principal component    </span>

 <span style="font-family: Comic Sans MS;font-size: 1.5em;color: #1A5F7A; background: inherit;">
<ul>
  <li>Advantages of PCA</li>
    <br />    <br />
        <ul>
      <li>PCA eliminates Multicollinearity, as all PC are perpendicular to each other and hence are independent.
</li>
                <br />    <br />
      <li>More features results in overfitting, PCA removes unnecessary information reducing overfitting.
</li>
    </ul>
        <br />    <br />
  <li>Limitations of PCA
        <br />    <br />
    <ul>
      <li>New features formed after PCA are not interpretable i.e. they are numbers in some directions.
</li>
            <br />    <br />
      <li>PCA tries to preserve the global shape of the data, the local structure is not preserved.
</li>
    </ul>
  </li>
</ul>

<span style="font-family: Comic Sans MS;font-size: 3em;color: #159895; background: inherit;">
    Implementation Manually
    </span>

```python
import numpy as np
from scipy import linalg as la
```

```python
data=np.array([
[6., 3., 2.],
[3., 2., 7.],
[5., 4., 2.],
[1., 4., 3.],
[7., 3., 1.0],
[5., 1., 8.],
[4., 2., 2.],
[8., 6., 6.],
[6., 3., 2.],
[7., 1., 1.]])
```

```python
# Center your data
dataNormed = (data - data.mean(axis=0))
dataNormed
```

```python
# Calculate the covariance matrix
cov = np.cov(dataNormed, rowvar=False) #If rowvar is True (default), then each row represents a variable.
cov
```

```python
# Calculate the eigenvalues and eigenvector of the covariance matrix:
evals, evecs = la.eig(cov)
```

```python
evals
```

```python
evecs
```

```python
# Sort the Eigen values and vector and select components
num_components=2
sorted_key = np.argsort(evals)[::-1][:num_components]
evals, evecs = evals[sorted_key], evecs[:, sorted_key]
```

```python
evals
```

```python
evecs
```

```python
# Multiply original data and Eigen vectors to get principal components
principal_components=np.dot(data,evecs)
print("Principal Components: \n", principal_components)
```

<span style="font-family: Comic Sans MS;font-size: 2em;color: #159895; background: inherit;">
    Our Parameters:
    </span>

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
X = np.empty((3, 7))

# Assign non-linear data to each column
X[:, 0] = np.array([1, 2, 3])
X[:, 1] = np.array([1, 4, 9])
X[:, 2] = np.array([0, 10, 20])
X[:, 3] = np.array([-1, -9, -14])
X[:, 4] = np.array([2, 2, 9])
X[:, 5] = np.array([7, 14, 28])
X[:, 6] = np.array([1, 6, 13])
X.shape = (7,3)
print(X)
```

```python
plt.plot(X)
plt.show()
```

```python
# n_components : Number of components to keep. if  not set (n_components == min(n_samples, n_features)
pca = PCA(n_components=1)
pca.fit_transform(X)
```

```python
pca = PCA()
pca.fit_transform(X)
```

```python
# components_ vectors are multiplied by the square root of n_samples and then divided by the singular values.
# ensure uncorrelated outputs (Whitening will remove some information)
pca = PCA(whiten = True)
pca.fit_transform(X)
```

```python
# svd_solver {auto, full, arpack, randomized}, default=auto
# if the input data is larger than 500x500 and the number of components lower than 80% of the data -- Use 'randomized'
# else exact full SVD is computed and optionally truncated afterward.
pca = PCA(svd_solver = 'auto')
pca.fit_transform(X)
```

```python
# run exact full SVD calling the standard LAPACK solver via scipy.linalg.svd
pca = PCA(svd_solver = 'full')
pca.fit_transform(X)
```

```python
# run SVD truncated to n_components calling ARPACK solver via scipy.sparse.linalg.svds.
# It requires strictly 0 < n_components < min(X.shape)
pca = PCA(svd_solver = 'arpack')
pca.fit_transform(X)
```

```python
# run randomized SVD by the method of Halko et al.
pca = PCA(svd_solver = 'randomized')
pca.fit_transform(X)
```

```python
# tol (default=0.0) Tolerance for singular values computed by svd_solver == arpack.
# Must be of range [0.0, infinity)
pca = PCA(svd_solver = 'arpack')
print(pca.fit_transform(X))
pca.singular_values_
```

```python
pca = PCA(svd_solver = 'arpack',tol = 0.2)
print(pca.fit_transform(X))
pca.singular_values_
```

```python
# iterated_power: default=auto Number of iterations for the power method
#  Must be of range [0, infinity).
pca = PCA(svd_solver = 'randomized',iterated_power =1)
print(pca.fit_transform(X))
```

```python
pca = PCA(svd_solver = 'randomized',iterated_power =2)
print(pca.fit_transform(X))
```

```python
# n_oversamples (default=10)
# the additional number of random vectors to sample
# ensure proper conditioning
pca = PCA(svd_solver = 'randomized',n_oversamples =2)
print(pca.fit_transform(X))
```

<span style="font-family: Comic Sans MS;font-size: 2em;color: #159895; background: inherit;">
    Our Attributes:

```python
pca = PCA()
pca.fit_transform(X)
```

```python
#  the directions of maximum variance in the data
c = pca.components_
print(pca.components_)
```

```python
# plot lines
plt.plot(c[:,0],label = "line 1")
plt.plot(c[:,1],label = "line 2")
plt.plot(c[:,2],label = "line 3")
plt.legend()
plt.show()
```

```python
# The amount of variance explained by each of the selected components
pca.explained_variance_
```

```python
# Percentage of variance explained by each of the selected components
pca.explained_variance_ratio_
```

```python
# The singular values corresponding to each of the selected components
pca.singular_values_
```

```python
# Per-feature empirical mean, estimated from the training set.
pca.mean_
```

```python
# The estimated number of components
pca.n_components_
```

```python
# Number of features in the training data. (Columns)
pca.n_features_in_
```

```python
# Number of samples in the training data. (Rows)
pca.n_samples_
```

```python
# The estimated noise covariance
pca.noise_variance_
```
